```python
opisy = []
for i in range(len(list(dict_txt.keys()))):
    all_opis = ""
    for key, value in dict_txt[list(dict_txt.keys())[i]].items():
        all_opis += f' {value}'
    opisy.append(all_opis)
```

```python
wyczyszczone_opisy = []
for el in opisy:
    if len(el) > 100:
        wyczyszczone_opisy.append(el)
    
```

```python
model = Top2Vec(wyczyszczone_opisy, embedding_model='universal-sentence-encoder-multilingual')
```

```python
model.get_num_topics()
```

```python
topic_words, word_scores, topic_nums = model.get_topics()
for topic in topic_nums:
    model.generate_topic_wordcloud(topic)
```

```python
for nazwa in translation:
    try:
        words, word_scores = model.similar_words(keywords=[nazwa], keywords_neg=[], num_words=20)
        for word, score in zip(words, word_scores):
            print(f"{nazwa} {word} {score}")
    except Exception as e:
        print(e)
```
